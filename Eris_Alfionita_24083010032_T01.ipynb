```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# range harga jual
x = np.linspace(3000, 10000, 500)

# rescale parameters
a = 6
loc = 3.5 * 1000
scale = 1.5 * 1000

# probability density, right tail skew
y = skewnorm.pdf(x, a=a, loc=loc, scale=scale) * 1000

np.random.seed(32) #NPM
y = y + np.random.normal(0, 0.005, size=x.shape)
y = y*1000

# harga produksi (biaya)
produksi_y = np.linspace(0, 1000, len(x))  # jumlah produksi (q)
produksi_z = np.linspace(3000, 2000, len(x)) + np.random.normal(0, 0.1, size=x.shape)*100
# biaya produksi per unit

# --- Perhitungan profit ---
# revenue = price * quantity
revenue = x * y

# cost = cost per unit * quantity
cost = produksi_z * produksi_y

# profit = revenue - cost
profit = revenue - cost

# cari indeks maksimum
idx_max = np.argmax(profit)
q_max = produksi_y[idx_max]
p_max = x[idx_max]
profit_max = profit[idx_max]

# --- Plot ---
plt.figure(figsize=(10,6))
plt.plot(produksi_y, profit, color="green", label="Profit")
plt.axvline(q_max, color="red", linestyle="--", label=f"q*={q_max:.0f}")
plt.scatter(q_max, profit_max, color="red", s=80, zorder=5)
plt.title("Profit vs Volume Produksi")
plt.xlabel("Volume Produksi (q)")
plt.ylabel("Profit")
plt.legend()
plt.grid(True)
plt.show()

print(f"Volume produksi optimal = {q_max:.0f} unit")
print(f"Harga jual optimal = Rp{p_max:,.0f}")
print(f"Keuntungan maksimum  Rp{profit_max:,.0f}")
```

Sumbu X = volume produksi (jumlah unit roti yang diproduksi).

Sumbu Y = profit (keuntungan dalam rupiah).

Dari grafik terlihat:
- Profit meningkat seiring bertambahnya produksi, tetapi hanya sampai di titik tertentu.
- **Titik maksimum terjadi pada volume produksi = 156 unit, dengan profit  Rp1.588.676.** Pada titik tersebut, harga jual optimal sekitar Rp4.094 per unit.
- Setelah melewati q = 156 unit, profit justru menurun tajam:
Hal ini mungkin karena biaya produksi terus naik, sementara pendapatan dari penjualan tidak lagi bertambah signifikan. Jika produksi terus ditambah, perusahaan bahkan bisa merugi (profit < 0).


Toko roti akan **memperoleh keuntungan paling besar bila memproduksi 156 roti dengan harga jual sekitar Rp4.094 per unit.**
Pada titik ini, **keuntungan maksimum yang bisa dicapai adalah Rp1.588.676.**

Artinya, jika produksi melebihi 156 unit, keuntungan justru menurun karena biaya produksi bertambah lebih cepat dibanding pendapatan penjualan.

**Kesimpulan :**
Strategi produksi optimal adalah dengan memproduksi sekitar 156 unit roti dan banyak memproduksi roti tidak berarti memperoleh profit/keuntungan lebih besar.
